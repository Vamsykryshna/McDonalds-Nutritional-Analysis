```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\VAMSI KRISHNA\OneDrive\Documents\Nutrical Dataset.csv"
data = pd.read_csv(file_path)

# Display the first few rows
print(data.head())
```

```python
# Check dataset information
print(data.info())

# Check for missing values
print(data.isnull().sum())
```

```python
data.fillna(0, inplace=True)
print("Missing values handled.")
```

```python
if data.duplicated().sum() > 0:
    data.drop_duplicates(inplace=True)
print("Duplicates removed.")
```

```python
print(data.describe())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Calories'], kde=True, bins=20, color='blue')
plt.title("Distribution of Calories")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=data, x='Category', y='Calories')
plt.title("Calorie Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Calories")
plt.xticks(rotation=45)
plt.show()
```

```python
top_items = data.nlargest(5, 'Calories')
print("Top 5 High-Calorie Items:\n", top_items[['Item', 'Calories']])
```

```python
clean_file_path = r"C:\Users\VAMSI KRISHNA\OneDrive\Documents\cleaned_mcdonalds_data.csv"
data.to_csv(clean_file_path, index=False)
print(f"Cleaned dataset saved at {clean_file_path}")
```
